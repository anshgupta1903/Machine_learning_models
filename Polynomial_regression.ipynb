```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate same synthetic data point or reproducability of result
np.random.seed(1)
```

```python
# Generate synthetic data points
X = 2*np.random.rand(100, 1)
```

```python
X
```

```python
# generate random synthetic y 
y = 4 + 3*X + 1.5*X**2 + np.random.randn(100,1)
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# perform polynomial regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree = 2, include_bias = False)
poly_features
```

```python
X_poly_train = poly_features.fit_transform(X_train)
```

```python
poly_features
```

```python
X_poly_train
```

```python
from sklearn.linear_model import LinearRegression
poly_reg = LinearRegression()
```

```python
poly_reg
```

```python
poly_reg.fit(X_poly_train, y_train)
```

```python
poly_reg.coef_
```

```python
poly_reg.intercept_
```

```python
# make predictions on the train data itself for the evaluation matrics
y_poly_predict = poly_reg.predict(X_poly_train)
```

```python
y_poly_predict
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, y_poly_predict)
```

```python
plt.scatter(X_train, y_train, label = 'train_data')
```

```python
degree = 2
plt.scatter(X_train, y_train, label = 'train_data')
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
plt.plot(X_range, poly_reg.predict(X_range_poly), color = 'red', label = f"Polynomial Regression (degree {degree})")
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```

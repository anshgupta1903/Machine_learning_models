```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the data
# here we are loading the data from sklearn
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
# Problem Statement: To Predect disease progression one year after baseline provided 10 features
# In one year how the sugar level has changed
```

```python
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names) # dataset has already been scaled down
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
# Data is already cleaned and preperation has already been done
```

```python
X = data[["bmi"]] # one independent variable
y = data['target'] # one dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
# train_test_split splits data randomly everytime you execute it train test data will be different 
# So to fix this we use random_state = some value
# random state when fixed helps us to reproduce the same result again and again
```

```python
X_train
```

```python
X_test
```

```python
# Scaling(optional) 
```

```python
# Model Training
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
X_train
```

```python
y_train
```

```python
# training of model
model.fit(X_train, y_train)
# at the backend it will use gradient descent to give optimal coefficient
```

```python
model.coef_ # attribute of the model
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# visualise the result
plt.scatter(X_test, y_test, color = 'black', label = "Actual Data")
plt.plot(X_test, y_pred, color = 'blue', linewidth = 3, label = "Predected LR line")
plt.xlabel("BMI")
plt.ylabel("One year progression target")
plt.title("Linear regrassion on diabetes data")
plt.legend()
plt.show()
```

```python
# Model evaluation
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
